```python
cat_feats=['purpose']
```

```python
final_data=pd.get_dummies(df,columns=cat_feats,drop_first=True)
```

```python
final_data.head()
```

```python
from sklearn.model_selection import  train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_data.drop('not.fully.paid',axis=1), final_data['not.fully.paid'], test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
prediction=dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,prediction))
```

```python
print(confusion_matrix(y_test,prediction))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rndom=RandomForestClassifier()
```

```python
rndom.fit(X_train,y_train)
```

```python
predictions=rndom.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
